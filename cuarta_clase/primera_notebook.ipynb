```python
try:
    number = 1
    string = "Hola"
    result = string + string
    var = string[10]
except TypeError as var2:
    print(var2)
    print("Hay un error de type error")
except IndexError as var2:
    print(var2)
    print("hay un error de index")
except:
    print("hay un error generico")
```

# Control Condiciones

```python
comp = 1 > 0
comp2 = 0 > 0

if (comp):
    print("cuando comp es TRUE")

if (comp):
    print("cuando comp es TRUE")
else:
    print("cuando comp es FALSE")

if (comp):
    print("cuando comp es TRUE")
elif (comp and comp2):
    print("otra condición pero negando la de arriba")
else: # Opcional
    print("Ninguna es TRUE")

if (comp):
    print("cuando comp es TRUE")
    if (comp2):
        print("cuando comp y comp2 son True")
else:
    print("cuando comp es FALSE")
    
if (comp and comp2):
    print("cuando comp y comp2 es TRUE")
```
